```python
# Import

import requests
import pandas as pd
import json
import pprint
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import geopandas as gpd
```

```python
# Load the GeoJSON file

path_object = Path('Resources/Bicycle_Thefts_Open_Data.geojson')
```

```python
# Convert the data to a Pandas dataframe

df = gpd.read_file(path_object)
df.head(10)
```

```python
df.info()
```

```python
# drop rows that have an NaN value 'bike cost' and columns ('LONG_WGS84', 'LAT_WGS84', 'HOOD_158', 'HOOD_140')

df_1 = df[df['BIKE_COST'].notna()]
df_new = df_1.drop(['HOOD_158', 'HOOD_140', 'geometry', 'OBJECTID'], axis=1)
df_new.info()
```

```python
df_new['REPORT_DATE']
```

```python
# convert the OCC_date and REPORT_DATE to datetime type
df_new['OCC_DATE'] = pd.to_datetime(df_new['OCC_DATE'])
df_new['REPORT_DATE'] = pd.to_datetime(df_new['REPORT_DATE'])
df_new['OCC_DATE'] = df_new['OCC_DATE'].dt.date
df_new['REPORT_DATE'] = df_new['REPORT_DATE'].dt.date
df_new.head()
```

```python
# LOAD to a SQlite database
from sqlalchemy import create_engine
```

```python
# Create Database Connection
# ----------------------------------
# Establish Connection

engine = create_engine("sqlite:///Resources/Bicycle_thefts.sqlite")
conn = engine.connect()
```

```python
df_new.to_sql("info", conn, if_exists="replace", index = False)
```

```python
data = engine.execute("SELECT * FROM info")

for record in data:
    print(record)
```
